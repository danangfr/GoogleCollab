```python
# --- Install dan Import Library ---
!pip install plotly seaborn --quiet

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import numpy as np

# Set style default lebih elegan
sns.set_theme(style="whitegrid")

# --- Data berdasarkan berita ---
pertumbuhan_dunia_2024 = 3.2
pertumbuhan_dunia_2025 = 2.9

kisaran_pertumbuhan_ri_2025 = (4.7, 5.5)
prakiraan_ri_2025 = (kisaran_pertumbuhan_ri_2025[0] + kisaran_pertumbuhan_ri_2025[1]) / 2 - 0.2

# --- Visualisasi ---

# 1. Bar Plot: Pertumbuhan Ekonomi Dunia
fig, ax = plt.subplots(figsize=(8,5))
tahun = ['2024', '2025']
pertumbuhan = [pertumbuhan_dunia_2024, pertumbuhan_dunia_2025]
colors = ['#00b4d8', '#ff6b6b']

sns.barplot(x=tahun, y=pertumbuhan, palette=colors, ax=ax)
ax.set_title('Pertumbuhan Ekonomi Dunia: 2024 vs 2025', fontsize=16, fontweight='bold')
ax.set_ylabel('Pertumbuhan (%)', fontsize=12)
for i, v in enumerate(pertumbuhan):
    ax.text(i, v + 0.05, f"{v}%", ha='center', fontweight='bold', fontsize=12)
plt.show()

# 2. Gauge Chart: Proyeksi Pertumbuhan Ekonomi Indonesia
fig = go.Figure(go.Indicator(
    mode = "gauge+number+delta",
    value = prakiraan_ri_2025,
    delta = {'reference': 5.1, 'increasing': {'color': "green"}, 'decreasing': {'color': "red"}},
    title = {'text': "Proyeksi Pertumbuhan Ekonomi Indonesia 2025"},
    gauge = {
        'axis': {'range': [4.0, 6.0]},
        'steps': [
            {'range': [4.0, 4.7], 'color': "#ffadad"},
            {'range': [4.7, 5.5], 'color': "#ffd6a5"},
            {'range': [5.5, 6.0], 'color': "#caffbf"}
        ],
        'threshold': {
            'line': {'color': "red", 'width': 4},
            'thickness': 0.75,
            'value': prakiraan_ri_2025
        }
    }
))
fig.update_layout(height=400)
fig.show()

# 3. Line Plot: Tren Volume Perdagangan Dunia
bulan = np.arange(1, 13)
volume_2024 = 100 - 0.5 * bulan  # tren datar sedikit menurun
volume_2025 = 100 - 1.5 * bulan  # tren menurun lebih cepat

plt.figure(figsize=(10,6))
plt.plot(bulan, volume_2024, label='2024', marker='o', color='#00b4d8', linewidth=2)
plt.plot(bulan, volume_2025, label='2025', marker='s', color='#ef476f', linestyle='--', linewidth=2)
plt.fill_between(bulan, volume_2025, volume_2024, color='mistyrose', alpha=0.4)
plt.title('Tren Volume Perdagangan Dunia: 2024 vs 2025', fontsize=16, fontweight='bold')
plt.xlabel('Bulan', fontsize=12)
plt.ylabel('Volume Indeks (Basis 100)', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# 4. Pie Chart: Aliran Modal Global 2025
labels = ['Aset Safe Haven (Emas, Yen, Euro)', 'US Treasury', 'Negara Berkembang']
sizes = [50, 20, 30]
colors = ['#06d6a0', '#118ab2', '#ffd166']

fig, ax = plt.subplots(figsize=(8,8))
wedges, texts, autotexts = ax.pie(sizes, labels=labels, colors=colors,
                                  autopct='%1.1f%%', startangle=140,
                                  textprops={'fontsize': 12})

ax.set_title('Arah Aliran Modal Global 2025', fontsize=16, fontweight='bold')
plt.setp(autotexts, size=12, weight="bold", color="black")
plt.show()
```

1.   **Pertumbuhan Ekonomi Dunia: 2024 vs 2025**: Pada 2025, ekonomi dunia diprediksi tumbuh lebih lambat di 2.9% dibandingkan 2024 sebesar 3.2%, akibat tekanan global dan pelemahan perdagangan internasional.

2.   **Proyeksi Pertumbuhan Ekonomi Indonesia 2025**: Pertumbuhan ekonomi Indonesia diproyeksikan mencapai sekitar 4.95% di tahun 2025, sedikit di bawah tengah kisaran target BI (4.7%-5.5%). Konsumsi domestik dan investasi akan tetap menjadi pendorong utama.

3. Tren Volume Perdagangan Dunia: 2024 vs 2025: Penurunan volume perdagangan global diperkirakan memburuk pada 2025, seiring dengan ketidakpastian ekonomi global dan melambatnya aktivitas ekspor-impor.

4. Arah Aliran Modal Global 2025: Penurunan volume perdagangan global diperkirakan memburuk pada 2025, seiring dengan ketidakpastian ekonomi global dan melambatnya aktivitas ekspor-impor.

5. Arah Aliran Modal Global 2025: Ketidakpastian global mendorong modal mengalir ke aset aman seperti emas, Yen, dan Euro, mengurangi porsi investasi di negara berkembang.
